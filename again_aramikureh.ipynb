```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.amazon.in/s?k=mobile+phone+under+20000&crid=2NUFT6GMXN6G&sprefix=%2Caps%2C515&ref=nb_sb_ss_recent_1_0_recent"
header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:142.0) Gecko/20100101 Firefox/142.0",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Referer": "https://www.amazon.in/",
    "DNT": "1",
    "Connection": "keep-alive",
}
```

```python
website = requests.get(url,headers=header)
```

```python
print(website)
```

```python
soup = BeautifulSoup(website.content,"html.parser")
```

```python
# Fetch links as List of Tag Objects
links = soup.find_all("a", attrs={'class':'a-link-normal s-no-outline'})
links
```

```python
link_list = [l.get('href') for l in links]
```

```python
d = {
    'title' : [],
    'price' : [],
    'rating' : [],
    'availability' :[]
}
```

```python
df = pd.DataFrame(d)
df
```

```python
link = link_list[2]
print(link)
```

```python

new_web = requests.get("https://www.amazon.com" + link, headers=header)
new_soup = BeautifulSoup(new_web.content, "html.parser")
```

```python
print(new_soup)
```

```python
for link in link_list:
    new_web = requests.get("https://www.amazon.com" + link, headers=header)
    new_soup = BeautifulSoup(new_web.content, "html.parser")
    title = new_soup.find("span", attrs={"id":'productTitle'}).text.strip()
    price = new_soup.find("span", attrs={"class":'a-price-whole'}).text.strip()
    rating = new_soup.find("span", attrs={"class":'a-icon-alt'}).text.strip()
    availability = new_soup.find("span", attrs={"class":'a-size-medium a-color-success'}).text.strip()
    d = {
        'title' : title,
        'price' : price,
        'rating' : rating,
        'availability' : availability
    }
    df = df.append(d, ignore_index=True)
```
